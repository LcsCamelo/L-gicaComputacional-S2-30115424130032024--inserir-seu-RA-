```python
print("P | Q | R | S | P → Q | R → S | P ∨ R | Q ∨ S | (P → Q) ∧ (R → S) | ((P → Q) ∧ (R → S) ∧ (P ∨ R)) | ((P → Q) ∧ (R → S) ∧ (P ∨ R)) → (Q ∨ S)|")
for a in (0,1):
    for b in (0,1):
        for c in (0,1):
            for d in (0,1):
                
                #Declarando as proposições principais
                if a==0:
                    P="V"
                else:
                    P="F"
                if b==0:
                    Q="V"
                else:
                    Q="F"
                if c==0:
                    R="V"
                else:
                    R="F"
                if d==0:
                    S="V"
                else:
                    S="F"
                
                #Condicional A = P → Q
                
                if P=="V" and Q=="F":
                    A = "F"
                else:
                    A = "V"
                
                #Condicional B = R → S
                
                if R=="V" and S=="F":
                    B = "F"
                else:
                    B = "V"
                    
                #Disjunção C = P ∨ R
                
                if P=="F" and R=="F":
                    C = "F"
                else:
                    C = "V"
                    
                #Disjunção D = (Q ∨ S)
                
                if Q=="F" and S=="F":
                    D = "F"
                else:
                    D = "V"
                    
                #Conjunção AB = (P → Q) ∧ (R → S)
                
                if A=="V" and B=="V":
                    AB = "V"
                else:
                    AB = "F"
                
                #Conjunção ABC = (P → Q) ∧ (R → S) ∧ (P ∨ R)
                
                if AB=="V" and C=="V":
                    ABC = "V"
                else:
                    ABC = "F"
                    
                #Condicional ABCD = ((P → Q) ∧ (R → S) ∧ (P ∨ R)) → (Q ∨ S)
                if ABC=="V" and D=="F":
                    ABCD = "F"
                else:
                    ABCD = "V"
                      
                print(f"{P} | {Q} | {R} | {S} |   {A}   |   {B}   |   {C}   |   {D}   |          {AB}         |                 {ABC}              |                    {ABCD}                  |")
 #TrabalhoFinalizado                      
                    
```
